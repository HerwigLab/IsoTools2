## Alternative splicing events
Isotools defines alternative spliceing events (ASE) with a graph based approach. 
The algorithm finds all ASE, classifies them by type (ES, IR, ME, 3AS, 5AS, TSS, PAS), and quantifies them with the PSI per sample.

The ASE can be used to calculate 2D embeddings (PCA, UMAP) representing the relation of the samples with respect to alternative splicing. 

([download link](https://oc-molgen.gnz.mpg.de/owncloud/s/gjG9EPiQwpRAyg3))

```python
from isotools import Transcriptome
import matplotlib.pyplot as plt

path='demonstration_dataset'
isoseq=Transcriptome.load(f'{path}/PacBio_isotools_substantial_isotools.pkl')
```

```python
splice_events=isoseq.alternative_splicing_events()
splice_events.splice_type.value_counts()
```

```python

#plot PCA embedding
from isotools.plots import plot_embedding
plt.rcParams["figure.figsize"] = (10,5)

pca={}

f,axs=plt.subplots(1,2)
plot_embedding(splice_events, ax=axs[0], labels=True, groups=isoseq.groups(), splice_types='all')
plot_embedding(splice_events, ax=axs[1], labels=True, groups=isoseq.groups(), splice_types='ES')
plt.tight_layout()


#f,axs=plt.subplots(4,2)
#for ax,t in zip(axs.flatten(),['all','3AS','5AS','ES','IR','ME', 'TSS', 'PAS']):
#    pca[t]=plot_embedding(splice_events, ax=ax, labels=True, groups=isoseq.groups(), splice_types=t)


#axs[0,0].legend(fontsize='medium', ncol=4,handleheight=2.4, labelspacing=0.05, bbox_to_anchor=(0, 1.1), loc='lower left')
#plt.tight_layout()

```
